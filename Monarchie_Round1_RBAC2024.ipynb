```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Import dataset
df= pd.read_csv(r'C:\Users\Admin\Downloads\loan_approval (Case Study B).csv')

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='cibil_score', hue='loan_status', kde=True, bins=20, palette='Set1', alpha=0.6)
plt.title('Distribution of CIBIL Scores by Loan Status')
plt.xlabel('CIBIL Score')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Import dataset
df= pd.read_csv(r'C:\Users\Admin\Downloads\loan_approval (Case Study B).csv')
plt.figure(figsize=(10, 6))


# Create a FacetGrid to split the data by loan status
g = sns.FacetGrid(df, col="loan_status", height=5, aspect=1.2)
g.map(plt.hist, "income_annum", bins=30, color="skyblue", edgecolor="black")

# Add titles and labels
g.set_axis_labels("Annual Income", "Frequency")
g.set_titles("Loan Status: {col_name}")
plt.suptitle("Annual Income Distribution by Loan Status", y=1.05)

# Show the plot
plt.show()
```
